```python
import seaborn as sns
import pandas as pd 
import cufflinks as cf
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
fig = px.line(y=tips['total_bill'])
fig.show()
```

```python
fig = px.bar(tips, x="day", y="tip", title="Tips by Day")
fig.show()
```

```python
fig = px.scatter(tips, x="total_bill", y="tip", title="Total Bill vs Tip")
fig.show()
```

```python
fig = px.box(tips, x="total_bill", y="day", title="Total Bill Distribution by Day")
fig.show()
```
