```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### We use distribution plots to visualize the distribution of quantitative data.

### Let's discuss some plots that allow us to visualize the distribution of a data set. These are:

#### 1. distplot/histplot
#### 2. joinplot
#### 3. pairplot
#### 4. rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique
```

```python
sns.displot(df['total_bill'], bins=100)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=100, kde=True)
```
