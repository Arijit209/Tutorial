```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

### Matlplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel('X-Axis')
plt.ylabel('Y-label')
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

### Object Oriented Way

##### Matplotlib allows the aspect ratio, DPI, and figure size to be specified when the Figure object is created. You can use the figsize and DPI keyword arguments.
##### Figsize is a tuple of the width and height of the figure in inches. DPI is the dots-per-inch (pixels per inch). 

```python
fig = plt.figure(figsize=(10,8), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])
axis2.plot(x,y)
```
