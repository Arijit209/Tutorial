```python
import numpy as np
import pandas as pd
```

#### Creating a DataFrame

```python
data = {
    'Name'   : ['Arijit', 'Pallabi', 'Anup', 'Indrani'],
    'Age'    : [38, 32, 75, 70],
    'City'   : ['Kolkata', 'Delhi', 'Chennai', 'Mumbai'],
    'Salary' : [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

```python
datalist = [
    ['Arijit', 38, 'Kolkata', 65000],
    ['Pallabi', 32, 'Delhi', 70000],
    ['Anup', 75, 'Chennai', 62000],
    ['Indrani', 70, 'Mumbai', 85000]
]
columns = ['Name', 'Age', 'City', 'Salary']
dflist = pd.DataFrame(datalist, columns=columns)
dflist
```

#### Selection and Indexing of Columns

```python
dflist['Name']
```

```python
dflist[['City', 'Salary']]
```

#### Creating a new column

```python
dflist['Designation'] = ['CA', 'Doctor', 'Engineer', 'Doctor']
dflist
```

#### Removing Columns

```python
dflist.drop(['Designation'], axis = 1, inplace=True)
```

```python
dflist
```

```python
dflist.drop([0], axis = 0)
```

#### Conditional Selection

```python
dflist
```

```python
# I only want to see those people whose age is above 40
dflist[dflist['Age'] > 40]
```

```python
# I only want to see those people whose age is above 40, and their city must be Mumbai
dflist[(dflist['Age'] > 40) & (dflist['City'] == 'Mumbai')]
```
