```python
import numpy as np
import pandas as pd
```

## Merging 2 DataFrames

```python
employees = pd.DataFrame({
    'employee_id' : [1, 2, 3, 4, 5],
    'name' : ['Arijit', 'Pallabi', 'Anup', 'Indrani', 'Aditish'],
    'department' : ['HR', 'IT', 'Finance', 'IT', 'HR']
})

salaries = pd.DataFrame({
    'employee_id' : [1, 2, 3, 6, 7],
    'salary' : [60000, 80000, 65000, 70000, 90000],
    'bonus' : [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)
```

```python
pd.merge(employees, salaries, on='employee_id') # on means common column name
```

```python
pd.merge(employees, salaries, on='employee_id', how='inner') # how inner means common of the two dataframes
```

```python
pd.merge(employees, salaries, on='employee_id', how='outer') # how outer means all of the two dataframes
```

```python
pd.merge(employees, salaries, on='employee_id', how='left') # how left means all of the data takes from left dataframe
```

```python
pd.merge(employees, salaries, on='employee_id', how='right') # how right means all of the data takes from right dataframe
```

## Concatination Of 2 DataFrames

```python
df1 = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2'],
    'B' : ['B0', 'B1', 'B2'],
    'C' : ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A' : ['A3', 'A4', 'A5'],
    'B' : ['B3', 'B4', 'B5'],
    'C' : ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

## Joing Of 2 DataFrames

```python
df1 = pd.DataFrame({
    'name' : ['Arijit', 'Pallabi', 'Anup', 'Indrani']
}, index=[1,2,3,4])
df2 = pd.DataFrame({
    'score' : [85, 90, 75, 80]
}, index=[2,3,4,5])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```

```python
df1.join(df2, how='inner')
```

```python
df1.join(df2, how='left')
```

```python
df1.join(df2, how='right')
```

```python
df2.join(df1)
```
