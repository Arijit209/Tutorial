# 📘 Array Operation

```python
import numpy as np
```

#### Arithmatic Operations

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
a1+a2 # shape should be the same, otherwise shows an error
```

```python
a1 - a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 // a2
```

```python
a1 % a2
```

```python
a1 ** a2
```

#### Broadcasting

```python
l = [10,20,30,40]
arr = np.array(l)
```

```python
arr + 10 # this is called braodcasting
```

```python
arr = np.arange(1, 26).reshape(5,5)
arr
```

```python
arr + 10
```

#### Deep And Shallow Copy

```python
arr = np.arange(1, 21)
arr
```

```python
# Shallow Copy
arr_slice = arr[:5]
arr_slice = arr_slice * 10
arr_slice
```

```python
arr
```

```python
arr1 = arr.copy()
arr1[0] = 10
arr1
```

```python
arr
```

```python
# Deep Copya
arr2 = arr
arr2[0] = 10
arr2
```

```python
arr
```

#### Stacking Array

```python
mat1
```

```python
mat2
```

```python
# In stack manipulation, only vector arrays work
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
np.vstack((a1, a2)) # vertical stack
```

```python
np.hstack((a1, a2)) # horizontal stack
```

```python
np.column_stack((a1, a2)) # column stack
```

#### Spliting Array

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.hsplit(c,2) # horizontal split but it needs equal division
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,2) # vertical split but it needs equal division
```

```python
np.vsplit(c,4)
```
