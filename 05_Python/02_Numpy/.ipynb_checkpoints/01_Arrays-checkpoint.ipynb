# 📘 NumPy / Array Learning Notes

```python
import numpy as np # import numpy library
```

#### Basic Array Creation

```python
arr = np.array([1,2,3,4,5]) # basic array creation
arr
```

```python
a = [1,2,3.5,4, 'hello'] # basic list creation
a
```

```python
np.array(a) # list convert to array
```

```python
l = [[1,2,3],[4,5,6],[7,8,9]]
l
```

```python
np.array(l) # two dimensinal array
```

#### Array Generation Function

```python
arr = np.arange(1,10) # range generation
arr
```

```python
arr = np.zeros(6) # zero generation
arr
```

```python
arr = np.zeros((4,8)) # 4 rows 8 columns
arr
```

```python
arr = np.ones((6, 6)) # generation ones
arr
```

```python
arr = np.linspace(1,5,3) # linear space (start, end, space)
arr
```

#### Random Generation Function

```python
arr = np.random.rand(5) # vectors and normalizations 0 to 5
arr
```

```python
arr = np.random.randn(5) # vectors and standardization -5 to 5
arr
```

```python
arr = np.random.randint(5) # single random number generate
arr
```

```python
arr = np.random.randint(10, 20, size=10) # integer array generate (low point, high point, number of array)
arr
```
