#### Array Generation Function

```python
arr = np.arange(1,10) # range generation
arr
```

```python
arr = np.zeros(6) # zero generation
arr
```

```python
arr = np.zeros((4,8)) # 4 rows 8 columns
arr
```

```python
arr = np.ones((6, 6)) # generation ones
arr
```

```python
arr = np.linspace(1,5,3) # linear space (start, end, space)
arr
```

#### Random Generation Function

```python
arr = np.random.rand(5) # vectors and normalizations 0 to 5
arr
```

```python
arr = np.random.randn(5) # vectors and standardization -5 to 5
arr
```

```python
arr = np.random.randint(5) # single random number generate
arr
```

```python
arr = np.random.randint(10, 20, 5) # integer array generate (low point, high point, number of array)
arr
```
