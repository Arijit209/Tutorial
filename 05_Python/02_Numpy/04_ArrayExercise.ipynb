#### Valid Sudoku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
b = np.sum(s, axis=1) # row axis for sum
# b==45
for i in b:
    if i != 45:
        print('Sudoku is not valid')
        break
else:
    print('Sudoku is valid for rows')
```

```python
c = np.sum(s, axis=0) # columns axis for sum
for i in c:
    if i != 45:
        print('Sudoku is not valid')
        break
else:
    print('Sudoku is valid for column')
```

```python
for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        # print(s[i:i+3, j:j+3])
        n = s[i:i+3, j:j+3]
        summ = n.sum()
        if summ != 45:
            print('Sudoku is not valid')
            break
else:
    print('Sudoku is valid for inner sum')
```

#### General Questions

```python
# Column: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
# Get the shape of the matrix
data.shape
```

```python
# Find the avarage age of the students
np.mean(data[:, 0])
```

```python
# Extract Math marks of all students
data[:, 1]
```

```python
# Find the highest Science mark
np.max(data[:,2])
```

```python
# Get details of the student who scored more than 90 in Math
data[data[:,1] > 90]
```

```python
# Increase Math marks of all students by 5
data
```

```python
data[:, 1] += 5
data
```

```python
# Find how many students are younger than 19
len(data[data[:,0] < 19])
```

```python
# Calculate the average marks in each subject (column-wise mean)
np.mean(data[:, 1:], axis=0)
```

```python
# Get data of the students who scored at least 80 in both subjects
data[(data[:, 1] >= 80) & (data[:, 2] >=80)]
```
