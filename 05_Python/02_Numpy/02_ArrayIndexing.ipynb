# 📘 NumPy / Array Indexing And Slicing Notes

```python
import numpy as np
```

#### Indexing And Slicing Vectors

```python
arr = np.arange(11,21)
arr
```

```python
# Indexing Part
arr_index = arr[9]
arr_index
```

```python
# Slicing Part
arr_slice = arr[2:]
arr_slice
```

```python
arr_slice = arr[:10]
arr_slice
```

```python
arr_slice = arr[2:10]
arr_slice
```

```python
arr_slice = arr[2:10:2] # slicing (start index, end index, steps)
arr_slice
```

#### Indexing And Slicing Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
# Indexing Part
arr_index = arr[5,4] # first row index then column index
arr_index
```

```python
# slicing Part
arr_slice = arr[0:2, 1:3] # first row indexes then column indexes
arr_slice
```

```python
arr_slice = arr[3:, 3:] # first row indexes then column indexes
arr_slice
```

```python
arr_slice = arr[:, 2] # first row indexes then column indexes
arr_slice
```

#### Boolean Indexing

```python
arr = np.arange(11, 21)
arr
```

```python
bool_index = arr%2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
